```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection details
host = "localhost"
port = 5432
username = "postgres"
password = "2003"
database = "guvi12"

# Create the engine string for PostgreSQL
engine_string = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)

# Define your DataFrames
# These should be replaced with your actual DataFrame creation logic
# Example data for demonstration purposes
cs = pd.DataFrame({'customer_id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
data_dictionary = pd.DataFrame({'column_name': ['customer_id', 'name'], 'description': ['ID of the customer', 'Name of the customer']})
exchange_rates = pd.DataFrame({'currency': ['USD', 'EUR'], 'rate': [1.0, 0.85]})
product = pd.DataFrame({'product_id': [101, 102], 'product_name': ['Widget', 'Gadget']})
sales = pd.DataFrame({'sale_id': [1, 2], 'amount': [150.00, 200.00]})
stores = pd.DataFrame({'store_id': [1, 2], 'location': ['New York', 'Los Angeles']})

# Define the table names and corresponding DataFrames
table_names = ['Customer', 'Datadictionary', 'Exchangerates', 'Product', 'Sales', 'Stores']
dataframes = [cs, data_dictionary, exchange_rates, product, sales, stores]

# Write each DataFrame to the corresponding PostgreSQL table
for table_name, df in zip(table_names, dataframes):
    df.to_sql(table_name, engine, if_exists="replace", index=False)

print("Success: Data has been written to the database.")
```
